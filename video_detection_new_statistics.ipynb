```python
import os
import matplotlib.pyplot as plt
import pandas as pd
from video_detect_by_erosion import process_video
```

```python
SAMPLES_FOLDER = "samples"
EROSION_VIDEO_FOLDER = "videos_detection_erosion"
ENHANCED_VIDEO_FOLDER = "videos_nofp_detection_erosion"
```

```python
results = {}

def get_sample_videos(folder=SAMPLES_FOLDER):
    """Retrieve original sample videos from the specified folder."""
    return [os.path.join(folder, f) for f in os.listdir(folder) if f.endswith(".mp4")]
```

```python
sample_video_paths = get_sample_videos()
print("Processing the following videos:", sample_video_paths)
```

```python
for video_path in sample_video_paths:
    video_name = os.path.basename(video_path)
    print(f"Processing {video_name}...")
    
    # Paths for processed videos
    erosion_video_path = os.path.join(EROSION_VIDEO_FOLDER, f"{video_name}_erosion.mp4")
    enhanced_video_path = os.path.join(ENHANCED_VIDEO_FOLDER, f"{video_name}_enhanced.mp4")
    
    # Process the video with erosion only
    results[video_name] = {}
    results[video_name]["erosion"] = process_video(video_path, erosion_video_path, use_erosion=True, save=True)
    
    # Process the video with erosion + false positive detection
    results[video_name]["enhanced"] = process_video(video_path, enhanced_video_path, use_erosion=False, save=True)
```

```python
data = []
for video_name, stats in results.items():
    data.append(
        {
            "video": video_name,
            "frames_total": stats["erosion"]["frames_numbers"],
            "detected_with_erosion": stats["erosion"]["frames_detected"],
            "false_detections_with_erosion": stats["erosion"]["false_detections"],
            "detected_enhanced": stats["enhanced"]["frames_detected"],
            "false_detections_enhanced": stats["enhanced"]["false_detections"],
        }
    )
```

```python
df = pd.DataFrame(data)
fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(25, 6))

x = range(len(df["video"]))
width = 0.35

# Bar plot of detection counts
ax1.bar(x, df["detected_with_erosion"], width, label="With Erosion", color="skyblue")
ax1.bar([i + width for i in x], df["detected_enhanced"], width, label="Enhanced", color="lightgreen")
ax1.set_ylabel("Frames with Detection")
ax1.set_title("AprilTag Detection Comparison")
ax1.set_xticks([i + width / 2 for i in x])
ax1.set_xticklabels(df["video"], rotation=45)
ax1.legend()

# Plot of detection rate
detection_rate_erosion = df["detected_with_erosion"] / df["frames_total"] * 100
detection_rate_enhanced = df["detected_enhanced"] / df["frames_total"] * 100

ax2.plot(df["video"], detection_rate_erosion, "o-", label="With Erosion", color="skyblue")
ax2.plot(df["video"], detection_rate_enhanced, "o-", label="Enhanced", color="lightgreen")
ax2.set_ylabel("Detection Rate (%)")
ax2.set_title("Detection Rate Comparison")
ax2.set_xticklabels(df["video"], rotation=45)
ax2.legend()
ax2.grid(True, linestyle="--", alpha=0.7)

# Bar plot of false detections
ax3.bar(x, df["false_detections_with_erosion"], width, label="With Erosion", color="blue")
ax3.bar([i + width for i in x], df["false_detections_enhanced"], width, label="Enhanced", color="green")
ax3.set_ylabel("False Detections")
ax3.set_title("False Detection Comparison")
ax3.set_xticks([i + width / 2 for i in x])
ax3.set_xticklabels(df["video"], rotation=45)
ax3.legend()

plt.tight_layout()
plt.show()
```
