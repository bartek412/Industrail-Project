```python
import os
import pandas as pd
import json
import numpy as np
```

```python
def compare_with_ground_truth(ground_truth_dir, examined_detections_dir, see_where=False):
    
    examined_detections_dir = os.getcwd()+examined_detections_dir
    ground_truth_dir = os.getcwd()+ground_truth_dir
    
    ground_truth = list(filter(lambda x: x[-5:] == '.json', os.listdir(ground_truth_dir)))
    examined_detections = list(filter(lambda x: x[-5:] == '.json', os.listdir(examined_detections_dir)))

    ground_truth.sort()
    examined_detections.sort()
    
    print(f"Ground_truth: {ground_truth}")
    print(f"Examined detections: {examined_detections}")
    
    data= 0
    ground_truth_data_json=0
    examined_detections_data_json=0
    ground_truth_frames=0
    df=0
    for g_t, e_d in zip(ground_truth, examined_detections):
        print(f"Comparing ground truth {g_t} with {e_d}")
        with open(examined_detections_dir + '/' + e_d, 'r') as file:
            examined_detections_data_json = json.load(file)
        with open(ground_truth_dir + '/' + g_t, 'r') as file:
            ground_truth_data_json = json.load(file)
        ground_truth_frames = list(ground_truth_data_json.keys())
        ground_truth_ids = [set(ground_truth_data_json[f]) for f in ground_truth_frames]
        examined_detections_ids = [set(examined_detections_data_json[f]) for f in ground_truth_frames]
        
        df = pd.DataFrame({
            "frame": ground_truth_frames,
            "ground_truth_ids": ground_truth_ids,
            "examined_detections_ids": examined_detections_ids
            })
    
        df['discrepancy'] = np.where((df['ground_truth_ids'] != df['examined_detections_ids']), 1, 0)
        print("\033[1m" + f"Discrepancy ratio (lower=better): {df['discrepancy'].sum(axis='index')/len(df.index):.4f}" + "\033[0m")
        if see_where:
            print(df.loc[df['discrepancy'] != 0])
            print('\n\n')
```
