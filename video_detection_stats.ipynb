```python
import os
from video_detect_by_erosion import process_video
```

```python
results = {}


def get_unprocessed_videos(folder="samples"):
    video_files = [
        f
        for f in os.listdir(folder)
        if f.endswith(".mp4") and "processed" not in f.lower()
    ]
    return [os.path.join(folder, f) for f in video_files]


video_paths = get_unprocessed_videos()
print("Videos to process:", video_paths)

# Process each video
for video_path in video_paths:
    print(video_path)
    results[video_path] = {}
    results[video_path]["eriosion"] = process_video(video_path, save=False)
    results[video_path]["noneriosion"] = process_video(
        video_path, use_erosion=False, save=False
    )
```

```python
results
```

```python
import matplotlib.pyplot as plt
import pandas as pd


data = []
for video_path, stats in results.items():
    video_name = os.path.basename(video_path)  # Get just the filename
    data.append(
        {
            "video": video_name,
            "frames_total": stats["eriosion"]["frames_numbers"],
            "detected_with_erosion": stats["eriosion"]["frames_detected"],
            "detected_without_erosion": stats["noneriosion"]["frames_detected"],
        }
    )

df = pd.DataFrame(data)


fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 6))


x = range(len(df["video"]))
width = 0.35

ax1.bar(x, df["detected_with_erosion"], width, label="With Erosion", color="skyblue")
ax1.bar(
    [i + width for i in x],
    df["detected_without_erosion"],
    width,
    label="Without Erosion",
    color="lightcoral",
)
ax1.set_ylabel("Frames with Detection")
ax1.set_title("AprilTag Detection Comparison")
ax1.set_xticks([i + width / 2 for i in x])
ax1.set_xticklabels(df["video"], rotation=45)
ax1.legend()


detection_rate_erosion = df["detected_with_erosion"] / df["frames_total"] * 100
detection_rate_no_erosion = df["detected_without_erosion"] / df["frames_total"] * 100

ax2.plot(
    df["video"], detection_rate_erosion, "o-", label="With Erosion", color="skyblue"
)
ax2.plot(
    df["video"],
    detection_rate_no_erosion,
    "o-",
    label="Without Erosion",
    color="lightcoral",
)
ax2.set_ylabel("Detection Rate (%)")
ax2.set_title("Detection Rate Comparison")
ax2.set_xticklabels(df["video"], rotation=45)
ax2.legend()
ax2.grid(True, linestyle="--", alpha=0.7)

plt.tight_layout()
plt.show()


print("\nSummary Statistics:")
print("-" * 50)
for _, row in df.iterrows():
    print(f"\nVideo: {row['video']}")
    print(f"Total Frames: {row['frames_total']}")
    print(
        f"Detection Rate with Erosion: {row['detected_with_erosion']/row['frames_total']*100:.1f}%"
    )
    print(
        f"Detection Rate without Erosion: {row['detected_without_erosion']/row['frames_total']*100:.1f}%"
    )
    print(
        f"Improvement with Erosion: {((row['detected_with_erosion'] - row['detected_without_erosion'])/row['frames_total']*100):.1f}%"
    )
```
