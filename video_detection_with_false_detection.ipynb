```python
import os
from video_detect_by_erosion import process_video
```

```python
results = {}


def get_unprocessed_videos(folder="samples"):
    video_files = [
        f
        for f in os.listdir(folder)
        if f.endswith(".mp4") and "processed" not in f.lower()
    ]
    return [os.path.join(folder, f) for f in video_files]


video_paths = get_unprocessed_videos()
print("Videos to process:", video_paths)
```

```python
for video_path in video_paths:
    print(video_path)
    results[video_path] = {}
    results[video_path]["erosion"] = process_video(video_path, save=False)
    results[video_path]["nonerosion"] = process_video(
        video_path, use_erosion=False, save=False
    )
```

```python
import matplotlib.pyplot as plt
import pandas as pd

data = []
for video_path, stats in results.items():
    video_name = os.path.basename(video_path)
    data.append(
        {
            "video": video_name,
            "frames_total": stats["erosion"]["frames_numbers"],
            "detected_with_erosion": stats["erosion"]["frames_detected"],
            "false_detections_with_erosion": stats["erosion"]["false_detections"],
            "frames_total_nonerosion": stats["nonerosion"]["frames_numbers"],
            "detected_without_erosion": stats["nonerosion"]["frames_detected"],
            "false_detections_without_erosion": stats["nonerosion"]["false_detections"],
        }
    )

df = pd.DataFrame(data)
```

```python
fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(20, 6))

x = range(len(df["video"]))
width = 0.35

# Bar plot of detection counts
ax1.bar(x, df["detected_with_erosion"], width, label="With Erosion", color="skyblue")
ax1.bar(
    [i + width for i in x],
    df["detected_without_erosion"],
    width,
    label="Without Erosion",
    color="lightcoral",
)
ax1.set_ylabel("Frames with Detection")
ax1.set_title("AprilTag Detection Comparison")
ax1.set_xticks([i + width / 2 for i in x])
ax1.set_xticklabels(df["video"], rotation=45)
ax1.legend()

# Plot of detection rate
detection_rate_erosion = df["detected_with_erosion"] / df["frames_total"] * 100
detection_rate_no_erosion = df["detected_without_erosion"] / df["frames_total"] * 100

ax2.plot(df["video"], detection_rate_erosion, "o-", label="With Erosion", color="skyblue")
ax2.plot(
    df["video"], detection_rate_no_erosion, "o-", label="Without Erosion", color="lightcoral"
)
ax2.set_ylabel("Detection Rate (%)")
ax2.set_title("Detection Rate Comparison")
ax2.set_xticklabels(df["video"], rotation=45)
ax2.legend()
ax2.grid(True, linestyle="--", alpha=0.7)

# Bar plot of false detections
ax3.bar(x, df["false_detections_with_erosion"], width, label="With Erosion", color="blue")
ax3.bar(
    [i + width for i in x],
    df["false_detections_without_erosion"],
    width,
    label="Without Erosion",
    color="red",
)
ax3.set_ylabel("False Detections")
ax3.set_title("False Detection Comparison")
ax3.set_xticks([i + width / 2 for i in x])
ax3.set_xticklabels(df["video"], rotation=45)
ax3.legend()

plt.tight_layout()
plt.show()
```
