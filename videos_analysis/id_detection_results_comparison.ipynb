```python
import os
import pandas as pd
import json
import numpy as np
```

```python
parent_dir = os.path.abspath(os.path.join(os.getcwd(), os.pardir))
videos_names = [video_name.replace('.mp4','') for video_name in os.listdir(parent_dir + '/videos')]
```

```python
videos_names
```

```python
results_path = parent_dir + '/videos_analysis/all_results.json'

with open(results_path, 'r') as file:
    results = json.load(file)

results_dict = {}

for r in results:
    for k,v in r.items():
        results_dict[k] = v

df = pd.DataFrame({
            "result_name": [],
            "tp": [],
            "tn": [],
            "fp": [],
            "fn": [],
            })
for result_name, result in results_dict.items():
    df.loc[len(df)] = [result_name,result["tp"],result["tn"],result["fp"],result["fn"]]
df
```

```python
for video_name in videos_names:
    print("----> VIDEO NAME:", video_name)
    print(df[df['result_name'].str.contains(video_name)])
```
